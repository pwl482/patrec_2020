```python
class LogisticRegression():
    """
    Creates an object for training a logistic regression model on a feature matrix and
    target variable.
    :param lr: Scalar for the stepsize of the gradient descent
    :type lr: float
    :param num_iter: Number of iterations the learning process is repeated.
    :type num_iter: int
    :param decision_threshold: Scalar as a cutoff for assigning labels based on predicted class probabilities
    :type decision_threshold: float
    """
    def __init__(self, lr=0.01, num_iter=10000, decision_threshold=0.5):
        self.loss = []
        self.w = None
        # NOTE: did deide not to use bias as individual parameter but include in X and self.w
        # self.b is thus commented out throught the following code
        #self.b = None
        self.lr =lr
        self.num_iter = num_iter
        self.decision_threshold = decision_threshold
        
    def sigmoid_function(self,X):
        """
        Implements the sigmoidal/logisitc function for some data X given slope/weight and bias of the class object
        """
        z = np.dot(X,self.w) 
        #+ self.b
        return 1/(1+np.exp(-z))
        
    def cross_entropy_loss(self, y_hat, y):
        """
        Calculates the mean cross entropy loss for a predictions y_hat of the sigmoidal function against the actual labels y.
        """
        return np.mean(-y*np.log(y_hat)-(1-y)*np.log(1-y_hat))
        
    def fit(self, X, y):
        """
        Find optimal fit of logistic function to input features X and corresponding labels y by 
        (1) maximizing log-likelihood for the training data given the estimated model parameters, and
        (2) minimizing the cross entropy loss of predicted labels against training labels
        """
        # Add bias/intercept into X
        X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
        # Weight and bias initialization
        self.w = np.zeros((X.shape[1]))
        #self.b = 0
        # Iterate: predict using current parameter estimation and update according to gradient
        for i in range(self.num_iter):
            y_hat = self.sigmoid_function(X)
            ## Document loss
            self.loss.append(self.cross_entropy_loss(y_hat, y))
            ## Update w based on predefined learning rate and observed gradient
            dth = 1/y.size * np.dot(X.T,(y_hat-y))
            self.w -= self.lr * dth
            #self.b -= self.lr * db

    def plot_loss(self):
        """
        Plots the mean cross entropy loss over the course of training iterations.
        """
        if len(self.loss) == 0:
            raise Exception("This Logistic Regression Object is not trained yet! Call the 'fit' function first!")
        else:
            plt.plot(list(range(len(self.loss))),self.loss)
            plt.xlabel('Iterations')
            plt.ylabel('Cross Entropy Loss')
            plt.show()

    def predict_proba(self,X):
        """
        Predicts label probabilities for input feature data.
        """
        if list(self.w) == None and self.b == None:
            raise Exception("This Logistic Regression Object is not trained yet! Call the 'fit' function first!")
        else:
            # Add bias to X
            X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
            return self.sigmoid_function(X)
        
    def predict(self, X):
        """
        Predict class labels for input data based on predicted label probabilities compared to decision threshold.
        The label with probability > decision_threshold is assigned
        """
        if list(self.w) == None and self.b == None:
            raise Exception("This Logistic Regression Object is not trained yet! Call the 'fit' function first!")
        else:
            # Add bias to X
            X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
            y = np.where(self.sigmoid_function(X) > self.decision_threshold, 1, 0)
        return y
    
    
    def accuray(self,yhat,y):
        """
        Compare predicted labels against ground truth and print accuracy
        """
        print('Accuracy: {:f}%'.format(yhat[np.where(yhat==y)].shape[0]/yhat.shape[0]))
        return None
```

```python
LR = LogisticRegression()
```

```python
y_train = y_train[np.logical_or(y_train==1,y_train==9)]
LR.fit(X_train, y_train)
```

```python
LR.plot_loss()
```

```python
yhat = LR.predict_proba(X_test)
```

```python
y_test = y_test[np.logical_or(y_test==1,y_test==9)]
LR.accuray(yhat,y_test)
```

Logistic Regression can be interpreted as a neural network with just a single layer. It uses the Cross Entropy to measure the performance of the layer (i.e. of the "trained" weight **w**). In ML we call this the **Loss function**.
