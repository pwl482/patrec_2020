
**Oja's algorithm**

```python
def oja(X, gamma):
    """
    Compute the first principal component for a given data set X using oja's algorithm.
    
    1. Initialize pc vectorw with length 1
    2. Start at random data point x and update w based on scaled distance of the projection of x on current w.
    3. Decrease learning rate by inverse iteration step
    4. Repeat 2.+3. for all data points
    
    Return  PC1 vector.
    """
    w = np.random.uniform(size=(X.shape[1]))
    # EA: doesn't w need to have length 1?
    w = w / np.linalg.norm(w, ord=2)
    np.random.shuffle(X)
    gamma = 1
    for i in range(X.shape[0]):
        x = np.array(X[i])
        # Update w
        w = w + (gamma*w)*x*(x-(w*x)*w)
        w = w / np.linalg.norm(w)
        # Update learning rate
        gamma = 1/(i+1)
    return(w)

def pca(X, gamma):
    """
    Retrieve the first two principal components for feature data X by applying oja's algorithm two times.
    
    1. Center the data
    2. Get PC1
    3. Normalize data with respect to PC1 by subtracting variance along the dimension of PC1.
    4. Get PC2
    5. Reduce dimensionality of X
    
    Return new coordinates of X.
    """
    X = X_features-np.mean(X_features, axis=0) # VB: i don't know if this is correct
    w1 = oja(X, gamma)
    X_new = X - (w1*X)*w1
    w2 = oja(X_new, gamma)
    W = np.array([w1, w2]).T # EA: tried to use arrays in both functions instead of matrices and data frames to have consistent handling
    # Dimensionality reduction (without data reconstruction)
    #Y = np.matmul(np.matrix(X), W)
    Y = X.dot(W) 
    return(Y)
        
```

```python
from sklearn.datasets import fetch_lfw_people
import numpy as np
from matplotlib import pyplot as plt


# #############################################################################
# Download the data, if not already on disk and load it as numpy arrays

lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4)

# introspect the images arrays to find the shapes (for plotting)
n_samples, h, w = lfw_people.images.shape

# for machine learning we use the 2 data directly (as relative pixel
# positions info is ignored by this model)
X = lfw_people.data
n_features = X.shape[1]

# the label to predict is the id of the person
y = lfw_people.target
target_names = lfw_people.target_names
n_classes = target_names.shape[0]

print("Total dataset size:")
print("n_samples: %d" % n_samples)
print("n_features: %d" % n_features)
print("n_classes: %d" % n_classes)
print("image_height: %d" % h)
print("image_width: %d" % w)


def show_faces(X):
    num_samples = 5
    indices = np.random.choice(range(len(X)), num_samples)
    print(indices.shape)
    sample_digits = X[indices]

    fig = plt.figure(figsize=(20, 6))

    for i in range(num_samples):
        ax = plt.subplot(3, 7.5, i + 1)
        img = sample_digits[i].reshape((h, w))
        plt.imshow(img, cmap='gray')
        plt.axis('off')
show_faces(X)
```
